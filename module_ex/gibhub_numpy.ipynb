```python
# numpy 총 17강 일괄 html로 저장하기 


from bs4 import BeautifulSoup
import requests

fdir = 'c:\\doit\\pdf\\'
save_name = 'numpy.html'
os.chdir(fdir)
ft_name = fdir + save_name
save_html = []
head_html = '<h1> https://076923.github.io/posts/#Python-Numpy </h1>'


for i in range(1, 18):
    url = f'https://076923.github.io/posts/Python-numpy-{i}/' # 숫자 1이 변하는 내용
    res = requests.get(url)
    html = res.text
    soup = BeautifulSoup(html, 'html.parser')
    result = soup.select('body > section > article > div')
    save_html.append(result[0])
    print(url)
    
#print(save_html)

with open( ft_name, 'w', encoding = 'utf-8') as f :
    f.write(head_html)
    f.write(str(save_html))
    
print('저장성공')
```
